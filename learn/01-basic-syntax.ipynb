## 1. 变量声明

Java 必须声明类型，Python 不需要

```python
# 说明：本段示例对应「1. 变量声明」，建议先看注释再运行，重点观察输出与预期是否一致。

# Java: String name = "kai";
#       int age = 18;
#       boolean isDev = true;

name = "kai"
age = 18
is_dev = True  # 注意：True/False 大写，不是 true/false

print(name, age, is_dev)
```

```python
# 说明：本段示例对应「1. 变量声明」，建议先看注释再运行，重点观察输出与预期是否一致。

# 动态类型：同一个变量可以换类型（Java 不行）
x = 10
print(type(x))  # <class 'int'>

x = "hello"
print(type(x))  # <class 'str'>
```

```python
# 说明：本段示例对应「1. 变量声明」，建议先看注释再运行，重点观察输出与预期是否一致。

# 类型注解（Type Hints）：不强制，但 FastAPI 大量使用
# Java: String greet(String name, int age) { ... }

# Q: 类型注解只用在函数参数吗？
# A: 不只是。函数参数、返回值、变量都可以加类型注解：
#    - 函数参数：def greet(name: str)
#    - 返回值：def greet() -> str
#    - 变量：name: str = "kai"（但变量注解很少用，因为 Python 能自动推断）
#    FastAPI 大量使用的主要是函数参数和返回值的注解

def greet(name: str, age: int) -> str:
    return f"{name} is {age}"

print(greet("kai", 18))

# 注意：类型注解只是提示，运行时不检查，传错类型也不会报错
# 静态检查靠工具：mypy 或 VSCode 的 Pylance 插件会在你写代码时实时标红，类似 TS 的效果，FastAPI 是个例外，它在运行时真的会读类型注解做参数校验，靠的是 pydantic 库
print(greet(123, "abc"))  # 不报错！Python 不管
```

## 2. 类型判断

```python
# 说明：本段示例对应「2. 类型判断」，建议先看注释再运行，重点观察输出与预期是否一致。

# Java: x instanceof String
# Python: isinstance(x, str)

# Q: str、int 这些类型从哪来的？需要 import 吗？
# A: 它们是 Python 的内置类型（builtins），和 Java 的 int/String 一样开箱即用，不需要 import
#    常见内置类型：str, int, float, bool, list, dict, set, tuple, None
#    它们同时也是构造函数：str(123) → "123"，int("42") → 42

x = "hello"
print(isinstance(x, str))    # True
print(isinstance(x, int))    # False
print(isinstance(x, (str, int)))  # True，可以同时判断多个类型
```

## 3. 字符串

```python
# 说明：本段示例对应「3. 字符串」，建议先看注释再运行，重点观察输出与预期是否一致。

# Q: f"..." 里的 f 是什么？
# A: f 是 format 的缩写，叫 f-string（格式化字符串字面量，Python 3.6+）
#    加了 f 前缀后，{} 里的内容会被当作 Python 表达式求值
#    不加 f 就是普通字符串，{} 不会被解析

name = "kai"
age = 18

# Java: String.format("My name is %s, age %d", name, age)
print(f"My name is {name}, age {age}")
print(f"1 + 1 = {1 + 1}")       # 大括号里可以写任意表达式
print(f"{'hello'.upper()}")      # 甚至可以调方法

# 对比：不加 f
print("My name is {name}")      # 输出字面量 {name}，不会替换
```

```python
# 说明：本段示例对应「3. 字符串」，建议先看注释再运行，重点观察输出与预期是否一致。

# Q: 三引号多行字符串运行时也是换行的吗？
# A: 是的，三引号保留原始的换行符，运行时也是多行的
#    和 Java 13+ 的 text block 效果一样

sql = """
SELECT *
FROM users
WHERE age > 18
"""
print(sql)       # 输出带换行的多行文本
print(repr(sql)) # 用 repr 可以看到实际的 \n 换行符

# 单引号和双引号完全等价（Java 里单引号是 char）
a = 'hello'
b = "hello"
print(a == b)  # True
```

```python
# 说明：本段示例对应「3. 字符串」，建议先看注释再运行，重点观察输出与预期是否一致。

# 常用字符串方法（和 Java 差别不大，注意命名风格是 snake_case）
s = "Hello World"

print(s.upper())          # HELLO WORLD
print(s.lower())          # hello world
print(s.split(" "))       # ['Hello', 'World']  返回 list，不是数组
print(s.replace("World", "Python"))
print(s.startswith("Hello"))
print("or" in s)          # True，用 in 判断包含（Java 用 contains）
print(len(s))             # 11，用 len() 不是 .length()
```

## 4. None（对比 Java 的 null）

```python
# 说明：本段示例对应「4. None（对比 Java 的 null）」，建议先看注释再运行，重点观察输出与预期是否一致。

# Java: if (x == null)
# Python: if x is None（用 is，不用 ==）

x = None
print(x is None)      # True
print(x is not None)  # False
print(type(x))        # <class 'NoneType'>

# 💡 实战：判断函数参数是否传值时永远用 is None，不用 == None
# 这是 Python 惯用法，也是代码审查里会被指出的点
# def process(data=None):
#     if data is None:   ← 正确写法
#         data = []
```

## 5. 运算符差异

```python
# 说明：本段示例对应「5. 运算符差异」，建议先看注释再运行，重点观察输出与预期是否一致。

# 整除和幂运算（Java 没有专门的运算符）
print(7 / 2)    # 3.5  普通除法（Java 里 7/2 = 3）
print(7 // 2)   # 3    整除
print(2 ** 10)  # 1024 幂运算（Java 要用 Math.pow）
```

```python
# 说明：本段示例对应「5. 运算符差异」，建议先看注释再运行，重点观察输出与预期是否一致。

# 逻辑运算符：用英文单词，不用符号
# Java: && || !
# Python: and or not

a = True
b = False

print(a and b)   # False（Java: a && b）
print(a or b)    # True （Java: a || b）
print(not a)     # False（Java: !a）
```

```python
# 说明：本段示例对应「5. 运算符差异」，建议先看注释再运行，重点观察输出与预期是否一致。

# is vs ==（重要！）
# ==  比较值是否相等（Java 的 .equals）
# is  比较是否是同一个对象（Java 的 ==）

a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)  # True，值相同
print(a is b)  # False，不是同一个对象
print(a is c)  # True，c 指向 a

# ⚠️ 坑：Python 对小整数（-5~256）和短字符串有缓存机制，
# 同样的值可能 is 返回 True，让人误以为 is 可以比较值
x = 256
y = 256
print(x is y)   # True（缓存范围内，碰巧是同一对象）
x = 1000
y = 1000
print(x is y)   # False（超出缓存范围）
# 结论：永远不要用 is 比较值，只用 is 判断 None / True / False
```

```python
# 说明：本段示例对应「5. 运算符差异」，建议先看注释再运行，重点观察输出与预期是否一致。

# in 运算符：判断元素是否在集合中（Java 没有，要用 contains）
print(3 in [1, 2, 3])        # True
print("h" in "hello")        # True
print("key" in {"key": 1})   # True，判断 key 是否在 dict 中
```

## 6. 代码块：缩进代替大括号

```python
# 说明：本段示例对应「6. 代码块：缩进代替大括号」，建议先看注释再运行，重点观察输出与预期是否一致。

# Java:
# if (age > 18) {
#     System.out.println("adult");
# } else {
#     System.out.println("minor");
# }

# Python：没有大括号，靠缩进，没有分号

# Q: 缩进必须4空格吗？2空格或3空格行不行？
# A: 语法上只要同一层级缩进一致就行（2、3、4都不报错）
#    但 PEP 8（Python 官方风格指南）规定用 4 空格，整个社区都遵守这个约定
#    VSCode 的 Python 插件默认就是 4 空格，不要改，跟着走就行

age = 20

if age > 18:
    print("adult")
else:
    print("minor")

# elif 对比 Java 的 else if
score = 85
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

```python
# 说明：演示基础循环写法，便于与你熟悉的 Java for 循环建立映射。

# for 循环
# Java: for (int i = 0; i < 5; i++)
# Python: for i in range(5)

# [Review] range(5) 不是列表，是惰性可迭代对象（类比 Java 的 IntStream.range）
# 不会一次性生成所有数字，而是按需产出，range(10000000) 也不占内存
for i in range(5):       # 产出 0,1,2,3,4
    print(i, end=" ")
    # Q: end 是什么？
    # A: end 是 print() 函数的关键字参数（keyword argument），控制打印后的结尾字符
    #    默认 end="\n"（换行），设为 end=" " 就用空格代替换行
    #    同理 sep 控制多个参数之间的分隔符，默认 sep=" "（空格）
    #    例：print("a", "b", sep="-") → "a-b"
    #    这不是 Python 关键字，而是 print 函数定义的参数名

print()  # 换行

# 直接遍历列表（Java 的 for-each）
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit, sep="-")
```

```python
# 说明：本段示例对应「6. 代码块：缩进代替大括号」，建议先看注释再运行，重点观察输出与预期是否一致。

# while 循环：和 Java 几乎一样，只是没有大括号
count = 0
while count < 3:
    print(count)
    count += 1
    # Q: Python 没有 count++ 和 count-- 吗？
    # A: 对，Python 没有 ++ 和 -- 一元运算符
    #    只能用 count += 1 和 count -= 1
    #    写 count++ 不会报语法错，但含义完全不同：
    #    Python 会把它解析成 count = +(+count)，即两次正号，值不变
```

## 7. 三元表达式

```python
# 说明：本段示例对应「7. 三元表达式」，建议先看注释再运行，重点观察输出与预期是否一致。

# Java: String result = age > 18 ? "adult" : "minor";
# Python: 语序不同！值在前，条件在中间

age = 20
result = "adult" if age > 18 else "minor"
print(result)
```

## 9. Truthy / Falsy 规则

Python 中以下值为 falsy（即 `if x` 判断为 False）：

```python
# 说明：本段示例对应「9. Truthy / Falsy 规则」，建议先看注释再运行，重点观察输出与预期是否一致。

# 这些都是 falsy
falsy_values = [False, 0, 0.0, "", [], {}, set(), None]

for val in falsy_values:
    print(f"{str(val):>10} -> {bool(val)}")

print()

# 实际用法：直接 if 判断，不用写 != null 或 .size() > 0
# Java: if (list != null && list.size() > 0)
# Python:
my_list = [1, 2, 3]
if my_list:  # 非空列表是 truthy
    print("list is not empty")

name = ""
if not name:  # 空字符串是 falsy
    print("name is empty")

# ⚠️ 坑：0、""、[] 都是 falsy，用 if x 做非空判断时会把"空但合法"的值误判掉
# 例：用户传来的 count=0，或空字符串 ""，都会被 if x 误判为"没传"
# 遇到这种情况要用 if x is not None 而不是 if x
def process(count=None):
    if count is None:        # ✅ 只判断"没传"
        count = 10
    if count:                # ❌ count=0 时也会走这里，逻辑错误
        print(f"count: {count}")
```

## 总结：关键差异速记

| 场景 | Java | Python |
|------|------|--------|
| 变量声明 | `String name = "kai";` | `name = "kai"` |
| 布尔值 | `true / false` | `True / False` |
| 空值 | `null` | `None` |
| **值比较** | `.equals()` | `==` |
| **引用比较（是否同一对象）** | `==` | `is` |
| 逻辑运算 | `&& \|\| !` | `and or not` |
| 代码块 | `{ }` | 缩进 |
| 分支 | `else if` | `elif` |
| 自增 | `i++` | `i += 1` |
| 三元 | `condition ? a : b` | `a if condition else b` |
| 长度 | `.length() / .size()` | `len()` |
| 包含 | `.contains()` | `in` |
| 格式化 | `String.format()` | `f"{}"` |
| 幂运算 | `Math.pow(2, 10)` | `2 ** 10` |
| 除法 | `7 / 2 = 3` | `7 / 2 = 3.5`，`7 // 2 = 3` |
